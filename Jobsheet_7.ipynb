```python
# -*- coding: utf-8 -*-
"""InventarisBarang - Exception Handling

Class InventarisBarang dengan manajemen stok barang menggunakan dictionary.
Menerapkan try-except-else-finally, raise exception, dan penanganan error.
"""

class InventarisBarang:
    """Kelas untuk mengelola inventaris barang dengan exception handling."""

    def __init__(self):
        """Konstruktor: inisialisasi dictionary protected _items."""
        self._items = {}  # dictionary: kode_barang -> jumlah_stok
        print("InventarisBarang siap digunakan.")

    def tambah_barang(self, kode, jumlah):
        """
        Menambahkan barang baru ke inventaris.

        Args:
            kode (str): Kode barang (unique key)
            jumlah (int): Stok awal (harus integer positif)

        Raises:
            KeyError: Jika kode barang sudah ada.
        """
        print(f"\n[tambah_barang] Mencoba menambah barang: {kode} = {jumlah}")

        # Validasi tipe dan nilai jumlah menggunakan try-except ValueError
        try:
            # Coba konversi ke int (jika input berupa string angka, misal "5")
            jumlah_int = int(jumlah)
            if jumlah_int <= 0:
                raise ValueError("Jumlah harus positif.")
        except ValueError as e:
            print(f"  ERROR (ValueError): {e}")
            return

        # Cek apakah kode sudah ada -> raise KeyError
        if kode in self._items:
            raise KeyError(f"Error: Kode barang '{kode}' sudah ada.")

        # Jika valid dan belum ada, tambahkan
        self._items[kode] = jumlah_int
        print(f"  -> Barang '{kode}' berhasil ditambahkan dengan stok {jumlah_int}.")

    def ambil_barang(self, kode, jumlah):
        """
        Mengurangi stok barang (pengambilan barang).

        Args:
            kode (str): Kode barang
            jumlah (int): Jumlah yang diambil (harus integer positif)

        Raises:
            KeyError: Jika kode barang tidak ditemukan.
            ValueError: Jika stok tidak mencukupi.
        """
        print(f"\n[ambil_barang] Mencoba mengambil {jumlah} unit barang '{kode}'")

        # Validasi jumlah
        try:
            jumlah_int = int(jumlah)
            if jumlah_int <= 0:
                raise ValueError("Jumlah yang diambil harus positif.")
        except ValueError as e:
            print(f"  ERROR (ValueError): {e}")
            return

        # Cek keberadaan kode (KeyError)
        try:
            stok_sekarang = self._items[kode]
        except KeyError:
            print(f"  ERROR (KeyError): Kode barang '{kode}' tidak ditemukan.")
            return

        # Cek kecukupan stok -> jika tidak cukup, raise ValueError
        if jumlah_int > stok_sekarang:
            raise ValueError(f"Error: Stok barang '{kode}' tidak mencukupi (tersedia: {stok_sekarang}, diminta: {jumlah_int}).")

        # Kurangi stok
        self._items[kode] -= jumlah_int
        print(f"  -> Pengambilan berhasil. Sisa stok '{kode}': {self._items[kode]}")

    def cek_stok(self, kode):
        """
        Mengecek stok barang tertentu.
        Menggunakan struktur try-except-else-finally.
        """
        print(f"\n[cek_stok] Mengecek stok untuk kode: '{kode}'")
        try:
            stok = self._items[kode]  # bisa memicu KeyError
        except KeyError:
            print(f"  -> Barang dengan kode '{kode}' TIDAK ditemukan.")
        else:
            # Dieksekusi hanya jika try sukses
            print(f"  -> Stok barang '{kode}' adalah: {stok} unit.")
        finally:
            # Selalu dieksekusi
            print(f"  [Pengecekan stok untuk '{kode}' selesai.]")

    def tampilkan_semua_stok(self):
        """Menampilkan semua barang yang ada di inventaris (utility)."""
        if not self._items:
            print("\n[Info] Inventaris kosong.")
        else:
            print("\n--- Daftar Semua Stok ---")
            for kode, stok in self._items.items():
                print(f"  {kode}: {stok} unit")


# ==================== KODE UTAMA ====================
if __name__ == "__main__":
    print("="*50)
    print("APLIKASI INVENTARIS BARANG")
    print("="*50)

    # 1. Buat instance InventarisBarang
    inv = InventarisBarang()

    # 2. Operasi tambah_barang (termasuk kasus kode sudah ada)
    print("\n--- Skenario tambah_barang ---")
    # Sukses
    inv.tambah_barang("B001", 100)
    inv.tambah_barang("B002", 50)

    # Coba tambah kode yang sudah ada (akan raise KeyError)
    try:
        inv.tambah_barang("B001", 30)
    except KeyError as e:
        print(f"  DITANGKAP DI KODE UTAMA: {e}")

    # Coba tambah dengan jumlah tidak valid (ValueError akan ditangani di dalam metode)
    inv.tambah_barang("B003", -5)   # ValueError internal, tidak raise ke atas
    inv.tambah_barang("B003", "sepuluh")  # ValueError konversi

    # 3. Operasi ambil_barang
    print("\n--- Skenario ambil_barang ---")
    # a. Kode tidak ada
    inv.ambil_barang("B999", 5)   # KeyError ditangani internal

    # b. Jumlah tidak valid
    inv.ambil_barang("B001", -10) # ValueError internal

    # c. Stok tidak cukup -> raise ValueError, harus ditangkap di kode utama
    try:
        inv.ambil_barang("B001", 200)  # stok hanya 100
    except ValueError as e:
        print(f"  DITANGKAP DI KODE UTAMA (ValueError): {e}")

    # d. Pengambilan sukses
    inv.ambil_barang("B001", 30)

    # 4. Operasi cek_stok (untuk kode ada dan tidak ada)
    print("\n--- Skenario cek_stok ---")
    inv.cek_stok("B001")   # ada
    inv.cek_stok("B002")   # ada
    inv.cek_stok("B999")   # tidak ada

    # Tampilkan semua stok akhir
    inv.tampilkan_semua_stok()
```
