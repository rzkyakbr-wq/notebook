```python
# Modul toko_hp.py berisi representasi kelas untuk Handphone
class Handphone:
    def __init__(self, merk, model, harga, stok):
        self.merk = merk
        self.model = model
        self.harga = harga
        self.stok = stok

    def tampilkan_info(self):
        print(f"HP: {self.merk} {self.model} | Harga: Rp {self.harga:,} | Stok: {self.stok} unit")

    def beli(self, jumlah):
        if jumlah <= self.stok:
            total_bayar = self.harga * jumlah
            self.stok -= jumlah
            print(f"Berhasil membeli {jumlah} unit {self.merk} {self.model}.")
            print(f"Total pembayaran: Rp {total_bayar:,}")
        else:
            print(f"Transaksi gagal! Stok {self.merk} {self.model} tidak mencukupi (Sisa: {self.stok}).")
```

```python
# File utama main.py untuk mengimpor modul toko_hp

def main():
    print("=== DAFTAR PRODUK AWAL ===")
    hp1 = Handphone("Samsung", "Galaxy S26", 15000000, 5)
    hp2 = Handphone("Apple", "iPhone 17", 18000000, 3)

    hp1.tampilkan_info()
    hp2.tampilkan_info()
    print("-" * 40)

    print("\n=== PROSES TRANSAKSI ===")
    # Mencoba membeli produk yang tersedia
    hp1.beli(2)

    # Mencoba membeli melebihi stok yang ada
    hp2.beli(5)
    print("-" * 40)

    print("\n=== DAFTAR PRODUK SETELAH TRANSAKSI ===")
    hp1.tampilkan_info()
    hp2.tampilkan_info()

if __name__ == "__main__":
    main()
```
